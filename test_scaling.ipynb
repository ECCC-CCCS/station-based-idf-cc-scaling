```python
import xclim
import numpy as np
import xarray as xr
```

```python
lat = np.random.uniform(low=-90, high=90, size=(4,5))
lon = np.random.uniform(low=-180, high=180, size=(4,5))
```

```python
test_arr = np.random.randint(1, 10, size=(4,5))
test_arr, lat , lon = np.around([test_arr, lat, lon], decimals=2)
test_da = xr.DataArray(data=test_arr,
                  coords={"lat": (["x","y"], lat),
                          "lon": (["x","y"], lon)},
                  dims=["x","y"])
test_da.attrs['units'] = 'mm/hour'
test_da
```

```python
delta_arr = np.random.randint(2, 4, size=(4,5))
delta_arr, lat , lon = np.around([delta_arr, lat, lon], decimals=2)
delta_da = xr.DataArray(data=delta_arr,
                  coords={"lat": (["x","y"], lat),
                          "lon": (["x","y"], lon)},
                  dims=["x","y"])
delta_da.attrs['units'] = 'delta_degreeC'
delta_da
```

```python
otpt = xclim.indices.clausius_clapeyron_scaled_precipitation(delta_da, test_da)
otpt
```

```python
basic_otpt = test_da * (1.07 ** delta_da)
basic_otpt
```

```python
otpt - basic_otpt
```
